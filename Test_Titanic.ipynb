```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
df = pd.read_csv('train.csv')

# Ver las primeras filas
print(df.head())

# Informacin general del dataset
print(df.info())

# Estadsticas bsicas
print(df.describe())

# Cuntas personas sobrevivieron?
print(df['Survived'].value_counts())

# Visualizacin: sobrevivientes por gnero
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Supervivencia por gnero')
plt.xlabel('Sobrevivi (0=No, 1=S)')
plt.ylabel('Cantidad de pasajeros')
plt.show()

# Promedio de edad por clase
print(df.groupby('Pclass')['Age'].mean())

# Visualizacin: edad vs supervivencia
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Edad vs Supervivencia')
plt.show()
```
